```python
# This is based on: https://github.com/JEddy92/TimeSeries_Seq2Seq/blob/master/notebooks/TS_Seq2Seq_Intro.ipynb
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

from scipy import stats

#Data is hourly
# df = pd.read_csv('out.csv', header=0)
# page_labels = df['rpair'].tolist()
# df = df.drop(df.columns[0:2],axis=1)

# df = df[:1000]

df = pd.read_csv('nov16.csv', header=0)
#page_labels = df['rpair'].tolist()
#df = df.drop(df.columns[0:2],axis=1)

#df = df.groupby(np.arange(len(df.columns))//4, axis = 1).mean()

#df = df[df.columns[:-1]]
#df = df[(np.abs(stats.zscore(df, axis=1)) < 2).all(axis=1)]

```

```python
df = df.T
```

```python
df = df.sample(frac=1)
```

```python
df = df.sample(frac=1)
```

```python
training_set = df.iloc[0:50,:]
```

```python
from sklearn.preprocessing import StandardScaler
# train the normalization
scaler = StandardScaler()
#scaler = scaler.fit(df)
# normalize the dataset and print
#df = scaler.transform(df)
#df = pd.DataFrame(df)
training_scaled = scaler.fit_transform(training_set)
training_scaled = pd.DataFrame(training_scaled)
```

```python
X_train = training_scaled.iloc[:,1:261]
y_train = training_scaled.iloc[:,261]

X_train = X_train.values
y_train = y_train.values
```

```python
# Reshaping
X_train = np.reshape(X_train, (50, 4, 65))
X_train, y_train = np.array(X_train), np.array(y_train)
X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 65))
```

```python
testing_set = df.iloc[50:61,:]
```

```python
scaler2 = StandardScaler()
#scaler = scaler.fit(df)
# normalize the dataset and print
#df = scaler.transform(df)
#df = pd.DataFrame(df)
testing_scaled = scaler2.fit_transform(testing_set)
testing_scaled = pd.DataFrame(testing_scaled)
```

```python
X_test = testing_scaled.iloc[:,1:261]
y_test = testing_scaled.iloc[:,261]

X_test = X_test.values
y_test = y_test.values
```

```python
# Reshaping
X_test = np.reshape(X_test, (11, 4, 65))
X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 65))
```

```python
import keras
from keras.preprocessing.text import one_hot, Tokenizer
from keras.preprocessing.sequence import pad_sequences
from keras.layers import *
from keras.layers.embeddings import Embedding
from keras.models import Sequential
```

```python
model = Sequential()
#model.add(Embedding(200, 100, input_length=261))
model.add(Bidirectional(LSTM(units=100, input_shape = (None, 65), return_sequences=True, dropout=0.2)))
model.add(Flatten())
model.add(Dense(1, activation='sigmoid'))
```

```python
model.compile(optimizer='adam', loss="mean_squared_error", metrics=['mse'])
```

```python
model.fit(X_train, y_train, batch_size=50, epochs=500, verbose=1)
```

```python
yhat = model.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error
# create empty table with 12 fields
test_predict = np.zeros(shape=(len(testing_set), 262) )
# put the predicted values in the right field
test_predict[:,261] = yhat.reshape(11,)
# inverse transform and then select the right field
trainPredict = scaler.inverse_transform(test_predict)[:,261]
```

```python
from sklearn.metrics import mean_squared_error
# calculate root mean squared error
#trainScore = math.sqrt(mean_squared_error(trainY[0], trainPredict[:,0]))
#print('Train Score: %.2f RMSE' % (trainScore))
import math
testScore = math.sqrt(mean_squared_error(df.iloc[50:61, 261].values, trainPredict))
print('Test Score: %.2f RMSE' % (testScore))
```
